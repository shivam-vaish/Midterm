## LOADING THE LIBRARIES

```python
from sklearn.datasets import load_diabetes
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
diabetes = load_diabetes()
```

```python
diabetes.target
```

```python
diabetes.data.shape
```

```python
Xtrain, Xtest, ytrain, ytest = train_test_split(diabetes.data,diabetes.target,test_size = 0.3, random_state =2)
```

## fitting the model

```python
reg = LinearRegression().fit(Xtrain, ytrain)
```

```python
reg.predict(Xtest)
```

```python
score =  reg.score(Xtest, ytest)
```

```python
print(score)
```

## SAVING THE MODEL

```python
import pickle
```

```python
with open("pickle_diabetes_model.pkl","wb") as file:
    pickle.dump(reg,file)
```

```python
# open the model
with open("pickle_diabetes_model.pkl","rb") as file:
    pickle_file = pickle.load(file)
```

```python
round(pickle_file.score(Xtest, ytest)*100,2)
```
